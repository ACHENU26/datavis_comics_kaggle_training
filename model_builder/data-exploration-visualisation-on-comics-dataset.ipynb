# DC COMIC EXPLORATION

## READ THE DATA

```python
# Read the file
nRowsRead = None

df1 = pd.read_csv('../data/dc-wikia-data.csv', delimiter=',', nrows = nRowsRead)

nRow, nCol = df1.shape

print(f'There are {nRow} rows and {nCol} columns')
df1.head()
```

## DATA IDENTYFICATION
---
* identifying data types and null values from each variables
* Dropping rows where null values are in Year and ALIVE variables or columns
* Identify null values in each variable before filling or replacing null values
* Converting and replacing missing values in YEAR column from float64 to int64
* Converting and replacing missing values with mean in APPEARANCES and from float64 to int64

```python
print('--- --- --- --- --- --- --- --- --- --- --- --- ')
print('')
print('dc-wikia-data.csv')
print('')
print('DATA TYPES')
print('')
print(df1.dtypes)
print('')
print('--- --- --- --- --- --- --- --- --- --- --- --- ')
print('')
df1_1=df1.dropna(subset=['YEAR'])
df1_1=df1_1.dropna(subset=['ALIVE'])
print('')
print("Before dropping null values: ",len(df1))
print("After dropping null values: ",len(df1_1))
print('Total ',len(df1)-len(df1_1),' rows are dropped')
print("Before updating null values : ",df1.shape,"After updating null values : ",df1_1.shape)
print('')
print('--- --- --- --- --- --- --- --- --- --- --- --- ')
print('')
print(df1_1.isnull().sum())
print('--- --- --- --- --- --- --- --- --- --- --- --- ')
```

## DATA MINING THEN DATA CLEANING
--- 
* Converting or filling missing values in ID, ALIGN, EYES, HAIR, SEX as Missing Value
* Deleting GSM column because it has lot of NaN
* Split FIRST APPEARANCE in Month and dropped or delete FIRST APPERANCE column
* Deleting urlslug column
* Converting or filling missing values in Month as December
* After fill null values and deleting GSM and FIRST APPEARANCE columns, checking how many number of Rows and Columns are their
* Verifying are they still any null values in dataframe
* Verifying data types of each variable
* Identifying unique values from related variables in data frame

```python
#df1_1['YEAR']=df1_1['YEAR'].replace("",0).fillna(0).astype(np.int64)
df1_1['YEAR']=df1_1['YEAR'].astype(np.int64)
#df1_12=dcwikia1[dcwikia1['APPEARANCES'].isnull()]
avg_appearances = df1_1['APPEARANCES'].mean()
df1_1['APPEARANCES']=df1_1['APPEARANCES'].fillna(avg_appearances)
#df1_1[dcwikia1['page_id']==289378]
df1_1['APPEARANCES']=df1_1['APPEARANCES'].astype(np.int64)
```

```python
# converted float64 to int64
print('--- --- --- --- --- --- --- --- --- --- --- --- ')
print('')
print(df1_1.dtypes)
print('')
print('--- --- --- --- --- --- --- --- --- --- --- --- ')


#Converting or filling missing values in ID, ALIGN, EYES, HAIR, SEX as Missing Value
df1_1['ID'] = df1_1['ID'].fillna('Missing values')
df1_1['ALIGN'] = df1_1['ALIGN'].fillna('Missing values')
df1_1['EYE'] = df1_1['EYE'].fillna('Missing values')
df1_1['HAIR'] = df1_1['HAIR'].fillna('Missing values')
df1_1['SEX'] = df1_1['SEX'].fillna('Missing values')

#Deleting GSM column because it has lot of NaN
del df1_1['GSM']

#Deleting urlslug column
del df1_1['urlslug']

#Split FIRST APPEARANCE in Month and dropped or delete FIRST APPERANCE column
fa=df1_1['FIRST APPEARANCE'].str.split(" ", n=1, expand=True)
df1_1['MONTH']=fa[1]

#Deleting FIRST APPEARANCE column
del df1_1['FIRST APPEARANCE']

#Converting month values
df1_1['MONTH']=df1_1['MONTH'].fillna('December')
df1_1['MONTH'] = np.where(df1_1['MONTH'] == 'January' ,1 ,np.where(df1_1['MONTH'] == 'February' ,2 ,np.where(df1_1['MONTH'] == 'March' ,3,np.where(df1_1['MONTH'] == 'April' ,4,np.where(df1_1['MONTH'] == 'May' ,5,np.where(df1_1['MONTH'] == 'June' ,6,np.where(df1_1['MONTH'] == 'July' ,7,np.where(df1_1['MONTH'] == 'August' ,8,np.where(df1_1['MONTH'] == 'September' ,9,np.where(df1_1['MONTH'] == 'October' ,10,np.where(df1_1['MONTH'] == 'November' ,11,np.where(df1_1['MONTH'] == 'December' ,12,0))))))))))))
```

```python
print('shape after data mining')
print('')
print('DATA TYPES')
print('')
print(df1_1.shape)
print('...............................................................................................')
print('null values count after data mining')
print('')
print(df1_1.isnull().sum())
print('...............................................................................................')
print('data types values after data mining')
print('')
print(df1_1.dtypes)
print('...............................................................................................')
print('nique values after data mining')
print('')
for i in df1_1[['ID','ALIGN','EYE','HAIR','SEX','ALIVE','YEAR','MONTH']]:
    print(i," : ",df1_1[i].unique())
    print("")
```
